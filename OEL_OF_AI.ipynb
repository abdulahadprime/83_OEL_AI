```python
import pandas as pd
```

```python
df = pd.read_csv("passenger_data.csv")
df
```

PART:1

```python
# 1.
total_passengers = len(df)
print("Total passengers:", total_passengers)
```

```python
# 2.
destination_counts = df['destination'].value_counts()
print("\nPassengers per destination:\n", destination_counts)
```

```python
# 3.
average_price_by_gender = df.groupby('gender')['ticket_price'].mean()
print("\nAverage ticket price by gender:\n", average_price_by_gender)
```

PART:2

```python
# 1.
def get_passengers_by_destination(dest):
    return df[df['destination'] == dest]['name'].tolist()

print("\nPassengers going to Karachi:", get_passengers_by_destination("Karachi"))
```

```python
# 2.
def get_highest_paying_passenger():
    highest = df[df['ticket_price'] == df['ticket_price'].max()]
    return highest[['name', 'destination']].iloc[0].to_dict()

print("\nHighest paying passenger:", get_highest_paying_passenger())
```

PART:3

```python
# 1.
sorted_df = df.sort_values(by='ticket_price', ascending=False)
print("\nSorted passengers by ticket price:\n", sorted_df[['name', 'ticket_price']])
```

```python
# 2.
top_3 = sorted_df[['name', 'age', 'ticket_price']].head(3)
print("\nTop 3 passengers:\n", top_3)
```
